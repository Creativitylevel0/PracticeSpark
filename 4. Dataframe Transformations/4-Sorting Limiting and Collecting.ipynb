####1. Read data from flight_time table

````python
flight_time_df = spark.table("dev.spark_db.flight_time")
display(flight_time_df)
````

####2. Find the top 3 most delayed flights on 2000-01-16 from AUS to ORD

Collect the results into a list and display the output as the following.
```
    AA flight delayed by 5.0 minutes
    AA flight delayed by 2.0 minutes
    UA flight delayed by 2.0 minutes
```

````python
from pyspark.sql.functions import col,expr

# (flight_time_df.filter((col("FL_DATE") == "2000-01-16") & (col("ORIGIN") == "AUS") & (col("DEST") == "ORD")).display())

top_3_df =  (
    flight_time_df
    .where("FL_DATE == '2000-01-16' and origin = 'AUS' and dest = 'ORD'") #we don't have to expr here because where expect it to be an expression
    .withColumn("Delayed_arrival", expr("arr_time - crs_arr_time"))
    .orderBy("Delayed_arrival", ascending=False)
    .limit(3)
)

top_3_list_of_rows = top_3_df.collect()
top_3_list_of_rows
````

2.2 Make a list of dictionary from a list of rows

````python
top_3_list = [row.asDict() for row in top_3_list_of_rows] # row is an pyspark.sql.Row it will help to convert it to a dictionary easily 
top_3_list 
````

2.3 Print the result in desired format

````python
for i in top_3_list:
    print(i["OP_CARRIER"]
            + " flight delayed by "
            + str(i["Delayed_arrival"].total_seconds()/60)
            + " minutes")
````

####2. Find the third most delayed flights on 2000-01-16 from AUS to ORD

````python
top_3rd_df =  (
    flight_time_df
    .where("FL_DATE == '2000-01-16' and origin = 'AUS' and dest = 'ORD'") #we don't have to expr here because where expect it to be an expression
    .withColumn("Delayed_arrival", expr("arr_time - crs_arr_time"))
    .orderBy("Delayed_arrival", ascending=False)
    .limit(3)
    .offset(2)
    )

top_3rd_df.display()

````
